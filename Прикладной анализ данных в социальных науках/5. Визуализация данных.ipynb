# 5. Визуализация данных

```python
import pandas as pd
import matplotlib.pyplot as plt

# df = pd.read_csv('PokemonData.csv', encoding='utf8')
df = pd.read_csv('data/PokemonData.csv', encoding='utf8')
df
```

```python
# df.HP.value_counts()
```

```python
# Постройте гистограмму распределения показателя здоровья (HP) у покемонов

plt.hist(df['HP'])
plt.show()
```

```python
# Атаки покемона

plt.hist(df['Attack'], alpha = 0.5, label="Обычная атака" )
plt.hist(df['SpAtk'], alpha = 0.5, label="Специальная атака")
plt.legend()

plt.xlabel('Мощность атаки')
plt.ylabel('Количество покемонов')

plt.show()
```

```python
# Правда ли, что чем покемон сильнее, тем больше его защита? 
# Постройте точечную диаграмму, чтобы проверить это.

plt.scatter(df['Attack'], df['Defense'], alpha = 0.3)
plt.xlabel('Мощность атаки')
plt.ylabel('Мощность защиты')
plt.show()
```

```python
# Визуализируем частоту наблюдений по типам покемонов.
variable = 'Type1'
df[variable].value_counts().plot(kind = 'bar')
```

```python
# Отразите распределение покемонов по типам и их легендарности. 
# В каких категориях присутствуют легендарные покемоны?

grouping_variable = 'Legendary'
categorical_variable = 'Type1'
# categorical_variable = 'Type2'
data = df.groupby(grouping_variable)[categorical_variable].value_counts()
data.unstack(0).plot(kind = 'bar', ylabel='frequency')
plt.xlabel("Тип покемонов")
plt.ylabel("Количество")
plt.title("Легендарные покемоны по типам в сравнении с обычными")
```
